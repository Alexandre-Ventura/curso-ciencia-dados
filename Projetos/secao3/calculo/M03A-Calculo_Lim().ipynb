```python
import matplotlib.pyplot as plt 
import numpy as np

# Funes que queremos analisar
def f1(x):
    return (x**2 - 4) / (x - 2)

def f2(x):
    return np.sin(x) / x

def f3(x):
    return x * np.cos(1/x)


# Valores para x
x = np.linspace(-10, 10, 500)

# Valor de c para os limites
c = 3

# Clculo dos limites usadno as funes definidas
limite_f1 = f1(c)
limite_f2 = f2(c)
limite_f3 = f3(c)

# Plotando os grficos das funes
plt.figure(figsize=(12,8))

plt.subplot(3,1,1)
plt.plot(x, f1(x))
plt.axvline(x=c, color='red', linestyle='--', label=f'Limite: {limite_f1}')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Grfico da Funo (x^2 - 4) / (x -2)')
plt.legend()

plt.subplot(3,1,2)
plt.plot(x, f2(x))
plt.axvline(x=c, color='red', linestyle='--', label=f'Limite: {limite_f2}')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Grfico da Funo sin(x8) / x')
plt.legend()

plt.subplot(3,1,3)
plt.plot(x, f3(x))
plt.axvline(x=c, color='red', linestyle='--', label=f'Limite: {limite_f3}')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Grfico da Funo x * cos(1/x)')
plt.legend()

plt.tight_layout()
plt.show()
```
